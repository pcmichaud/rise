# Fit Exploration

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
sim = pd.read_csv('../output/simulation/simulated_reference.csv',sep=',',header=None)
sim.columns = ['id','age','wealth','work','claim','income','uninsured','ame',
               'cons','dead','fhealth','health','oop','medexp','value','tr']
```

```python
sim['work'] = sim['work'].replace(2,0)
sim['uninsured'] = sim['uninsured'].replace(2,0).replace(3,0)
sim['claim'] = sim['claim'].replace(1,0).replace(2,1)
sim['vgood'] = sim['health'].replace(2,0).replace(3,0).replace(4,1)
sim['poor'] = sim['health'].replace(2,1).replace(3,0).replace(4,0)
sim['dead'] = sim['dead'].replace(1,0).replace(2,1)
```

```python
age = sim.groupby(['age']).mean()
```

```python
vars = ['wealth','work','claim','income','uninsured','ame',
               'cons','dead','poor','vgood','oop','medexp','value','tr']
for v in vars:
    plt.figure()
    plt.plot(age.index,age[v],label=v)
    plt.xlabel('age')
    plt.ylabel('outcome '+v)
    plt.title(v)
    plt.show()
```
